# 1. Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
rank = []
video_name = []
artist = []
upload_date = []
view = []
```

```python
def sol1(url):
    
    #loading web driver
    driver = webdriver.Chrome(r'C:\Users\Rahul\Downloads\chromedriver_win32\chromedriver.exe')
    driver.maximize_window()
    
    #getting url
    driver.get(url)
    
    #finding and clicking on the right link
    click1 = driver.find_element_by_xpath("//a[@class='external text']")
    click1.click()
    
    #checking and clicking on 1st link
    click2 = driver.find_element_by_xpath("//div[@class='mw-search-result-heading']/a")
    print(click2.get_attribute('href'))
    click2.click()
    
    #checking table where the details are present
    table1 = driver.find_element_by_xpath("//div[@class='mw-parser-output']/table[1]/tbody")
#     print(table1.text)
    
    #ranks
    rank_tag = driver.find_elements_by_xpath("//div[@class='mw-parser-output']/table[1]/tbody/tr/td[1]")
    for i in rank_tag:
        rank.append(i.text)
    time.sleep(1)
        
    #Video name
    video_name_tag = driver.find_elements_by_xpath("//div[@class='mw-parser-output']/table[1]/tbody/tr/td[2]")
    for i in video_name_tag:
        video_name.append(i.text)
    time.sleep(1)
    
    #Artist
    artist_tag = driver.find_elements_by_xpath("//div[@class='mw-parser-output']/table[1]/tbody/tr/td[3]")
    for i in artist_tag:
        artist.append(i.text)
    time.sleep(1)
    
    #upload_date
    upload_date_tag = driver.find_elements_by_xpath("//div[@class='mw-parser-output']/table[1]/tbody/tr/td[5]")
    for i in upload_date_tag:
        upload_date.append(i.text)
    time.sleep(1)
    
    #views
    views_tag = driver.find_elements_by_xpath("//div[@class='mw-parser-output']/table[1]/tbody/tr/td[4]")
    for i in views_tag:
        view.append(i.text)
    time.sleep(1)
    
```

```python
sol1('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/')
```

```python
sol1_data = pd.DataFrame({
    'Rank':rank,
    'Name':video_name,
    'Artist':artist,
    'Upload date':upload_date,
    'Views (in billions)':view
})
    
sol1_data
```

```python
sol1_data.to_csv('sol1_data.csv')
```

```python
def sol5(url):
    driver = webdriver.Chrome(r'C:\Users\Rahul\Downloads\chromedriver_win32\chromedriver.exe')
    driver.maximize_window()
    driver.get(url)
    
    #clicking explore tab
    expl_tag = driver.find_element_by_xpath("//div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
    expl_tag.click()
    expl_tag.click()
    
    time.sleep(2)
    #trnd
    trend_tag = driver.find_element_by_xpath("//div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
    trend_tag.click()
    

    #language used:
    lang_tag = driver.find_elements_by_xpath("//article[@class='Box-row']/div[2]")
    for i in lang_tag:
        try:
            language.append((i.find_element_by_xpath(".//span[@class='d-inline-block ml-0 mr-3']")).text)
        except NoSuchElementException:
            language.append('-')

    #repository
    rep_tag = driver.find_elements_by_xpath("//article[@class='Box-row']/h1")
    for i in rep_tag:
        rep.append(i.text.replace('\n',''))

    #description
    desc_tag = driver.find_elements_by_xpath("//article[@class='Box-row']")
    for i in desc_tag:
        try:
            about.append((i.find_element_by_xpath(".//p[@class='col-9 color-text-secondary my-1 pr-4']")).text)
        except NoSuchElementException:
            about.append('-')
            
    #contributers
    count_tag = driver.find_elements_by_xpath("//article[@class='Box-row']")
    for i in count_tag:
        try:
            count.append((i.find_element_by_xpath(".//div[@class='f6 color-text-secondary mt-2']/a[2]")).text)
        except NoSuchElementException:
            count.append('-')
        
#     for i in urls[0:3]:
#         driver.get(i)
```

```python
def sol7(url):
    driver = webdriver.Chrome(r'C:\Users\Rahul\Downloads\chromedriver_win32\chromedriver.exe')
    driver.maximize_window()
    driver.get(url)
    
    #getting link of recruiters
    recruiters_tag = driver.find_element_by_xpath("//ul[@class='midSec menu']/li[2]/a")
    driver.get(recruiters_tag.get_attribute('href'))
    
    #search and click for data scientist
    driver.find_element_by_xpath("//input[@class='sugInp']").send_keys('Data science')
    driver.find_element_by_xpath("//button[@id='qsbFormBtn']").click()
    
    for i in range(0,2):
        
        #name_tag
        name_tag = driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
        for i in name_tag:
            name.append(i.text)
        
        #designation
        desig_tag = driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
        for i in desig_tag:
            designation.append(i.text)
            
        #company
        comapny_tag = driver.find_elements_by_xpath("//p[@class='highlightable']/a[2]")
        for i in comapny_tag:
            company.append(i.text)
            
        #skills
        skill_tag = driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
        for i in skill_tag:
            skills.append(i.text)
            
        #location_tag
        highlight_table_tag = driver.find_elements_by_xpath("//p[@class='highlightable']")
#     print(len(highlight_table_tag))
        for i in highlight_table_tag:
            try:
                location.append((i.find_element_by_xpath(".//small[@class='ellipsis']")).text)
            except NoSuchElementException:
                location.append('-')


        next_but_tag = driver.find_elements_by_xpath("//div[@class='pagination']/button")
        for i in next_but_tag:
            if i.text == "Next":
                i.click()
```

# 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the ShowAllDataset page through code.

```python
dataset_name = []
data_type = []
task = []
attribute_type = []
no_of_instances = []
no_of_attributes = []
year_note = []
```

```python
def sol10(url):
    driver = webdriver.Chrome(r'C:\Users\Rahul\Downloads\chromedriver_win32\chromedriver.exe')
    driver.maximize_window()
    driver.get(url)
    
    time.sleep(2)
    #clicking show all data
    driver.find_element_by_xpath("//span[@class='normal']/b/a").click()
    time.sleep(2)
    
    #dataset_name
    dataset_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in dataset_tag:
        dataset_name.append(i.text)
    time.sleep(2)
    
    #data_type
    data_type_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
    for i in data_type_tag[1:]:
        data_type.append(i.text)
    time.sleep(2)
     
    #task
    task_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
    for i in task_tag[1:]:
        task.append(i.text)
    time.sleep(2)

    #attribute
    attr_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
    for i in attr_tag[1:]:
        attribute_type.append(i.text)
    time.sleep(2)
    
    #no_instances
    instances_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
    for i in instances_tag[1:]:
        no_of_instances.append(i.text)
    time.sleep(2)
    
    #attributes
    attr_no_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
    for i in attr_no_tag[1:]:
        no_of_attributes.append(i.text)
    time.sleep(2)
    
    #year_note
    year_tag = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
    for i in year_tag[1:]:
        year_note.append(i.text)
```

```python
sol10('https://archive.ics.uci.edu/ml/index.php')
```

```python
sol10_data = pd.DataFrame({
    'dataset':dataset_name,
    'data_type':data_type,
    'task':task,
    'attribute_type':attribute_type,
    'instances':no_of_instances,
    'no of attribute':no_of_attributes,
    'year_note':year_note
})

sol10_data
```
